```python
def concat_dok( Ds, func = sum, args = [],):
    from scipy.sparse import dok_matrix
    if func == max:
        args = []
    elif func == sum:
        args = [()]
    
    same = 1
    s0 = Ds[0].shape
    keys = set()
    for D in Ds:        
        same *= D.shape == s0
        assert same,'size checking failed'
        keys = keys | set(D.keys())
    
        
    OUTPUT = dok_matrix( Ds[0].shape )
    it = keys

    d = dict()
    c = counter([],INF = 1,per = 10000)
    for k in it:
#         v = sum( (D.get(k,(0,)) for D in Ds),() )
        v = func( (D.get(k,(0,)) for D in Ds), *args )
        c.count()
#         v1 = D_curr[x,y]
        d[k] = v
    OUTPUT.update(d)
    c.summary("concatenating sparse matrix")
    return OUTPUT

def matrify( lst, flat = False, l = None):
    if not l:
        l = max(itertools.izip(*lst))
        if isinstance(l,tuple):
            l = l[0]
#     l = self.l
    OUTPUT = dok_matrix( (l + 1, l + 1), dtype = 'int')
    it = (tuple(x) for x in lst)
    count = Counter(it)
    if not flat:
        for k,v in count.iteritems():
            count[k] = (v,)
    OUTPUT.update(count)
    return OUTPUT
```

```python
# mapper = list2dict(header)



# fname = full('$SEQlib/crosshits.v4_1_0.tsv')
def import_crosshit(fname = full('$SEQlib/crosshits.v4_1_0.tsv'),
                   debug = 0):
    def parser(line, header = None, domain2node = None):
        lst = line.rstrip('\n').split('\t')
        d = dict( zip(header,lst) )

        node1 = domain2node[d['model_id' ] ]
        node2 = domain2node[d['domain_id'] ]

        nodes = [node1,node2]
        nodes = sorted(nodes)

        jdict = {
             'node1_id'    : nodes[0],
             'node2_id'    : nodes[1],   
             'prc_evalue'  : d['evalue'],
             'prc_overlap' : d['pc_overlap'],
             'hmm_overlap' : d['hmm_max_overlap'],
             'ssap_score'  : d['ssap_score'],
             'ssap_align'  : d['num_align'],
             'prc_align'  :  d['pc_align'],
             'prc_seqid'  :  d['pc_seq_id'],
             'rmsd'  :       d['rmsd'],
            }
         
         
        return jdict
    domain2node = dict(domain.objects.values_list('domain_id','classification'))
    behave = "Importing crosshits from %s" % fname
    with open(fname,'r') as f:
        lines = f.readlines()
    #     lines = open(fname,'r').readlines()
        header = None
        OUTPUT = []
        c = counter([],INF=1,per = 10000,
                   behave = behave)
        for line in lines:
            line = line.rstrip('\n')
            if not header:
                header = line.split('\t')
            else:
                c.count()
                if debug:
                    jdict = parser(line, header, domain2node)
                try:
                    jdict = parser(line, header, domain2node)
                    OUTPUT += [jdict]
#                 except KeyError:
#                     pass
                except Exception as e:
                    c.fail(None)
    c.summary()
    return OUTPUT
# buf = import_crosshit(
#     full('$SEQlib/crosshits.v4_1_0.tsv'),
#     debug = 0)
```
